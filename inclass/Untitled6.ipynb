## 가로로 합치기 merge(): 행의 선택은?

```python
# 중간고사 데이터 만들기
test1 = pd.DataFrame({'id'      : [1, 2, 3, 4, 5],
                      'midterm' : [60, 80, 70, 90, 85]})

# 기말고사 데이터 만들기
test2 = pd.DataFrame({'id'    : [1, 2, 3, 4, 5],
                      'final' : [70, 83, 65, 95, 80]})
```

```python
test1
```

```python
test2  #비슷하지만 어느정도 다른 데이터가 존재할 수 있다--> 합쳐야하는 경우가 존재하기 때문에 알아두자
```

```python
total = pd.merge(test1, test2, how = 'left', on = 'id')  #how ='inner'.on='id', 양쪽에 동일하게 있는 것만 가지고 합함(inner방식), 공통 기준은 id
total
```

# 세로로 합치기: concat()

```python

group_a = pd.DataFrame({'id'   : [1, 2, 3, 4, 5],
                        'test' : [60, 80, 70, 90, 85]})

group_b = pd.DataFrame({'id'   : [6, 7, 8, 9, 10],
                        'test' : [70, 83, 65, 95, 80]})
```

```python

group_a
```

```python
group_b
```

```python
# 데이터 합쳐서 group_all에 할당
group_all = pd.concat([group_a, group_b]) #인덱스가 그대로 붙여짐 바꿀수 있는 방법은?-  reset_index()메서드 사용하기
group_all
```

##심화학습 책에 없음

공식 매뉴얼 문서에 존재함

코드를 이해해야지 좋다.

```python
df1 = pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df1
```

```python
df2
```

```python
# key 및 rkey 열에 df1 및 df2를 병합, 공통 키가 다른 경우
# 값 열에는 기본 접미사인 _x 및 _y가 추가
df1.merge(df2, left_on='lkey', right_on='rkey') # =pd.merge(df1,df2,left~~)
# value_x,y같은 경우 칼럼명이 동일하기 때문에 다른걸로 인식하기 위해서 만들어짐
```

```python
#공통인것만 뺌 foo bar baz존재---> foo에 1,5/5,8 공통값 존재
```

```python
 # 겹치는 열에 지정된 왼쪽 및 오른쪽 접미사를 추가하여 DataFrames df1 및 df2를 병합 .
df1.merge(df2, left_on='lkey', right_on='rkey', suffixes=('_left', '_right'))
```

```python
# 이름을 주는 것도 내가 지정할 수 있음 (위에거 참고, 값은 똑같지 변수명이 달라짐)
```
